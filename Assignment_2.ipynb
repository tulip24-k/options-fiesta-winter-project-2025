•Strike Price (K=100)

• Premium (P=15) that you pay (for buying) or receive (for selling), if buying the premium
will be positive and for selling it will be negative.

• Arange of Spot Prices (for example, from K −50 to K +30)

• Type of option (Call or Put).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# taking inputs
K = 100   # Strike price
P = 15    # Premium  
S = np.arange(K-30,K+50)# Spot price range from 50 to 130
```

```python
# Payoff calculations
buy_call = np.maximum(0, S - K) - P
sell_call = P - np.maximum(0, S - K)
buy_put = np.maximum(0, K - S) - P
sell_put = P - np.maximum(0, K - S)
```

```python
# Plotting(4 SubPlots)

fig,axs = plt.subplots(2, 2, figsize=(12, 8))

# Buy Call
axs[0, 0].plot(S, buy_call, color='green')
axs[0, 0].axhline(0, color='black')
axs[0, 0].axvline(K, linestyle='--', color='gray')
axs[0, 0].set_title("Buy Call Option")
axs[0, 0].set_xlabel("Spot Price")
axs[0, 0].set_ylabel("Profit / Loss")
axs[0, 0].grid(True)

# Sell Call
axs[0, 1].plot(S, sell_call, color='red')
axs[0, 1].axhline(0, color='black')
axs[0, 1].axvline(K, linestyle='--', color='gray')
axs[0, 1].set_title("Sell Call Option")
axs[0, 1].set_xlabel("Spot Price")
axs[0, 1].set_ylabel("Profit / Loss")
axs[0, 1].grid(True)

# Buy Put
axs[1, 0].plot(S, buy_put, color='blue')
axs[1, 0].axhline(0, color='black')
axs[1, 0].axvline(K, linestyle='--', color='gray')
axs[1, 0].set_title("Buy Put Option")
axs[1, 0].set_xlabel("Spot Price")
axs[1, 0].set_ylabel("Profit / Loss")
axs[1, 0].grid(True)

# Sell Put
axs[1, 1].plot(S, sell_put, color='orange')
axs[1, 1].axhline(0, color='black')
axs[1, 1].axvline(K, linestyle='--', color='gray')
axs[1, 1].set_title("Sell Put Option")
axs[1, 1].set_xlabel("Spot Price")
axs[1, 1].set_ylabel("Profit / Loss")
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))

plt.plot(S, buy_call, label="Buy Call", color="green")
plt.plot(S, sell_call, label="Sell Call", color="red")
plt.plot(S, buy_put, label="Buy Put", color="blue")
plt.plot(S, sell_put, label="Sell Put", color="orange")
# Reference lines
plt.axhline(0, color='black', linewidth=1)
plt.axvline(K, color='gray', linestyle='--', label="Strike Price")
# Labels and title
plt.xlabel("Spot Price at Expiry")
plt.ylabel("Profit / Loss (₹)")
plt.title("Option Payoff Diagrams (K=100, Premium=15)")
plt.legend()
plt.grid(True)

plt.show()
```

1. Delta – Price Sensitivity to Spot Price

Delta measures how much the option price changes when spot price changes by ₹1.

Range:
Call Option Delta: 0 to +1
Put Option Delta: -1 to 0

If:
Call option Delta = 0.6
Spot price increases from 100 → 101
Then:
Premium increases by ≈ ₹0.60
*BUYER BENEFITS FROM THE RISE OF PRICE BECAUSE HIGHER THE SPOT PRICE EASIER IT IS TO TOUCH THE STRIKE PRICE .THEREFORE HE INCREASES THE PREMIUM AS A COMPENSATION FOR THAT. *

Similarly:
Put option Delta might be -0.4
If spot price increases by ₹1
Premium decreases by ₹0.40
*SELLER BENEFITS FROM THE INCREASE OF SPOT PRICE,AS IT IS A PUT OPTION .SO HE DECREASES THE PREMIUM TO MAKE A FAIR DEAL. *

2. Gamma (Γ) – Rate of Change of Delta

Gamma measures how fast Delta changes when spot price changes.

Range:
Always positive for both calls and puts

Highest for At-The-Money (ATM) options
High Gamma means Delta is unstable
Near expiry and ATM options have high Gamma.
*MAY BE THE INTUITION IS THAT WHEN GAMMA INCREASES,THE BUYER CAN PROFIT IN EITHER WAYS.SO HE INCREASES THE PREMIUM AS THE BUYER CAN MAKE PROFIT. *

Example:
 Call Delta = 0.6
Gamma = 0.05
If spot price increases by ₹1:
New Delta = 0.65

 3. Theta (Θ) – Time Decay

Theta measures how much option value decreases per day due to time passing.

Range:
Negative for buyers(*TIME VALUE GOES ON DECREASING SO THE LATER THE BUYER TRADES OFF,THE MORE TIME VALUE HE LOSES *)
Positive for sellers(*AS TIME MOVES ON ,THE SELLERS ENJOY THE MORE PART OF THE PREMIUM *)

Example:
 Theta = -0.4
Each day, option loses ₹0.40 even if spot price doesn’t move

Sellers benefit from Theta

4. Vega (ν) – Sensitivity to Volatility

Vega measures how much option price changes when volatility changes by 1%.

Range:
Positive for both calls and puts

Highest for ATM options

Example:
Vega = 0.8
If volatility increases by 1%
Option price increases by ₹0.80

Higher volatility → higher option value
*THE SELLER INCREASES THE PREMIUM AS A COMPENSATION OF THE RISK HE IS TAKING *
5. Rho (ρ) – Interest Rate Sensitivity

Rho measures change in option price due to 1% change in interest rate.

Range:
Call Rho: Positive
Put Rho: Negative

Generally small for short-term options and beneficial for long term options
