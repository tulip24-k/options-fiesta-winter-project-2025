```python
import numpy as np
import matplotlib.pyplot as plt
# taking input form the user
K = float(input("Enter strike price K: "))
P = float(input("Enter premium P: "))
spot_start = float(input("Enter the starting spot price: "))
spot_end = float(input("Enter the ending spot price: "))
option= input("Enter option type (call/put): ").lower()

S = np.linspace(spot_start, spot_end, 100)

#define payoff functions
if option == "call":
    if P > 0:
        payoff = np.maximum(0, S - K) - P   # Buy Call
        strategy = "Buy Call"
    else:
        payoff = P - np.maximum(0, S - K)   # Sell Call
        strategy = "Sell Call"

elif option == "put":
    if P > 0:
        payoff = np.maximum(0, K - S) - P   # Buy Put
        strategy = "Buy Put"
    else:
        payoff = P - np.maximum(0, K - S)   # Sell Put
        strategy = "Sell Put"

#create the graphs
plt.figure(figsize=(8, 5))
plt.plot(S, payoff, label=strategy, linewidth=2)

plt.axhline(0)                 # Zero Profit line
plt.axvline(K, linestyle="--") # Strike Price

plt.xlabel("Spot Price at Expiry")
plt.ylabel("Profit / Loss")
plt.title(f"Profit/Loss vs Spot Price ({strategy})")
plt.legend()
plt.grid(True)

plt.show()

```
